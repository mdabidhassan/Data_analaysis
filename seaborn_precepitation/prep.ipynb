```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rain_data=pd.read_csv('average_precipitation.csv')
```

1. Bar Plot â€“ Comparing precipitation across years:
2. Use: When comparing quantities across categories.
Best for: Discrete data, especially categorical data (e.g., sales by region, counts of events).
Example: Comparing average precipitation across different years or months.
Explanation: Bar plots are ideal for comparing a single numeric value (like average) across different categories (like different years or regions).
When to Use:

When you have categorical data and want to compare individual category values.
For datasets where categories have no inherent ordering (like regions or product types).


```python
plt.figure(figsize=(14, 7))
sns.barplot(
    x='Year', 
    y='Average Precipitation (mm)', 
    data=rain_data, 
    palette="coolwarm",  # Custom color palette
)
plt.xticks(rotation=45, fontsize=10)  # Rotate x-axis labels for better readability
plt.title("Average Precipitation per Year (2001â€“2021)", fontsize=16, weight='bold')  # Add title
plt.xlabel("Year", fontsize=12, weight='bold')  # Add x-axis label
plt.ylabel("Average Precipitation (mm)", fontsize=12, weight='bold')  # Add y-axis label
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines for clarity
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

**3. Scatter Plot â€“ Showing the relationship between years and precipitation:

```python
y = np.sin(x)

# Plot the sine wave
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="sin(x)", color="b", linewidth=2)
plt.title("Sinusoidal Plot (sin(x))", fontsize=16, weight='bold')
plt.xlabel("X-axis (radians)", fontsize=12, weight='bold')
plt.ylabel("Amplitude", fontsize=12, weight='bold')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python

x_sin = np.linspace(0, 2 * np.pi, len(rain_data))
y_sin = np.sin(x_sin)

# Set the style and color palette for the plot
sns.set_style("white")
sns.set_palette("husl")

# Create the plot
plt.figure(figsize=(14, 7))

# Plot precipitation data as bars
sns.barplot(x='Year', y='Average Precipitation (mm)', data=rain_data, color="skyblue", alpha=0.7)

# Overlay the sine wave
plt.plot(rain_data['Year'], y_sin * (max(rain_data['Average Precipitation (mm)']) - min(rain_data['Average Precipitation (mm)'])) + min(rain_data['Average Precipitation (mm)']),
         label="Sinusoidal Curve", color="red", linewidth=2)

# Titles and labels
plt.title("Average Precipitation with Sinusoidal Overlay", fontsize=16, weight='bold')
plt.xlabel("Year", fontsize=12, weight='bold')
plt.ylabel("Precipitation (mm)", fontsize=12, weight='bold')

# Grid and legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.xticks(rotation=45)
plt.tight_layout()
```

**10. Facet Grid / Subplot â€“ Showing precipitation trends across different categories (if you had categorical data like regions or months):
10. Facet Grid / Subplot
Use: To create multiple subplots based on a categorical variable.
Best for: Comparing the same plot across different categories.
Example: Comparing precipitation trends across different regions.
Explanation: Facet grids allow you to create multiple plots, each corresponding to a subset of the data based on a categorical variable.
When to Use:

When you want to split a dataset by categories and show the same plot for each category.
For comparing different segments or categories in your dataset.

```python
# Adding a 'Region' column to the data for demonstration purposes
rain_data['Region'] = ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 
                       'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
                       'North', 'South', 'East', 'West', 'North']

# Plotting Facet Grid
g = sns.FacetGrid(rain_data, col="Region", height=4, aspect=1.5)
g.map(sns.lineplot, "Year", "Average Precipitation (mm)")
g.set_titles("{col_name} Region")
g.set_axis_labels("Year", "Precipitation (mm)")
g.tight_layout()
plt.show()
```

6. Pie Chart
Use: To show proportions of a whole.
Best for: Showing percentage breakdowns of categorical data.
Example: Showing the proportion of precipitation in different months of the year.
Explanation: Pie charts are used to show relative proportions and help highlight which categories make up a whole.
When to Use:

When you need to show proportions or relative percentages of a total.
Best for small datasets with a few categories (e.g., market share by brand).


```python
# Calculate the total precipitation
total_precipitation = rain_data['Average Precipitation (mm)'].sum()

# Calculate proportions for each year
rain_data['Proportion'] = rain_data['Average Precipitation (mm)'] / total_precipitation

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(rain_data['Proportion'], labels=rain_data['Year'], autopct='%1.1f%%', startangle=90)
plt.title("Proportion of Precipitation by Year")
plt.show()

```
